```python
# Create an empty DataFrame to store the results
pDF = pd.DataFrame(columns=['pid', 'average_popularity'])

# Iterate through each playlist
for i in range(len(playlist_data['playlists'])):
    playlist = playlist_data['playlists'][i]
    
    # Initialize variables to calculate average popularity
    total_popularity = 0
    num_tracks = 0
    
    # Iterate through each track in the playlist
    for track in playlist['tracks']:
        track_uri = track['track_uri'].split(":")[2]
        
        # Check if track URI exists in song data
        if track_uri in song_data:
            song = song_data[track_uri]
            total_popularity += song['popularity']
            num_tracks += 1
    
    # Calculate average popularity for the playlist
    if num_tracks > 0:
        average_popularity = total_popularity / num_tracks
    else:
        average_popularity = 0  # Default value if no tracks in the playlist
    
    # Add a new row to the DataFrame with playlist ID and average popularity
    pDF.loc[i] = {'pid': playlist['pid'], 'average_popularity': average_popularity}

# Print the first few rows of the new DataFrame
print(pDF.head())
```
